## Preparación de los datos

### Selección

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from tpot import TPOTClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
csv = 'dataset/weatherAUS.csv'
df_weatherAUS = pd.read_csv(csv,sep=",")
```

```python
df_weatherAUS.head()
```

### Limpieza

```python
#df_weatherAUS = df_weatherAUS.dropna()
```

### Construcción

```python
df_weatherAUS['WindSpeedMax'] = df_weatherAUS[['WindSpeed9am','WindSpeed3pm']].max(axis=1)
df_weatherAUS['WindSpeedMin'] = df_weatherAUS[['WindSpeed9am','WindSpeed3pm']].min(axis=1)
df_weatherAUS['WindSpeedMean'] = df_weatherAUS[['WindSpeed9am','WindSpeed3pm']].mean(axis=1)
```

### Integración

```python
# df weatherAUS_Locations
csvL = 'dataset/weatherAUS_Locations.csv'
df_weatherAUS_Locations = pd.read_csv(csvL,sep=",")
```

```python
# Fusionar por la característica 'Location'
df_weatherAUS = pd.merge(df_weatherAUS, df_weatherAUS_Locations[['Location','Latitud','Longitud','Is_Coastal']], on='Location')
```

```python
df_weatherAUS.head()
```

```python
df_weatherAUS.describe().T
```

```python
df_weatherAUS.info()
```

### Formatear datos

```python
df_weatherAUS['Date']  = pd.to_datetime(df_weatherAUS['Date'])
df_weatherAUS['Year'] = df_weatherAUS['Date'].dt.year
df_weatherAUS['Month'] = df_weatherAUS['Date'].dt.month
df_weatherAUS['Day'] = df_weatherAUS['Date'].dt.day
df_weatherAUS['Week'] =df_weatherAUS['Date'].dt.isocalendar().week
```

```python

le = LabelEncoder()
df_weatherAUS['LocationEncoder'] = le.fit_transform(df_weatherAUS['Location'])
df_weatherAUS['WindDir9amEncoder'] = le.fit_transform(df_weatherAUS['WindDir9am'])
df_weatherAUS['WindDir3pmEncoder'] = le.fit_transform(df_weatherAUS['WindDir3pm'])
df_weatherAUS['WindGustDirEncoder'] = le.fit_transform(df_weatherAUS['WindGustDir'])
df_weatherAUS['Is_CoastalEncoder'] = le.fit_transform(df_weatherAUS['Is_Coastal'])
df_weatherAUS['RainTomorrowEncoder'] = le.fit_transform(df_weatherAUS['RainTomorrow'])
df_weatherAUS['RainTodayEncoder'] = le.fit_transform(df_weatherAUS['RainToday'])
```

### Guardar nuevos DF

```python

# Crear copia 
dfLocations = df_weatherAUS.copy()

# Lista de columnas categóricas que deseas eliminar
columnas_categoricas = ['Location', 'WindDir9am','WindDir3pm','WindGustDir','Is_Coastal','Date','RainTomorrow','RainToday']

# Eliminar las columnas categóricas
dfLocations = dfLocations.drop(columnas_categoricas, axis=1)


# Separar en ciudades costeras y no costeras
ciudades_si_costeras = dfLocations.query('Is_CoastalEncoder == 1')
ciudades_no_costeras = dfLocations.query('Is_CoastalEncoder == 0')

ciudades_si_costeras.to_csv('./dataset/weatherAUS_LocationsCiudadesSiCosteras.csv',index=False)
ciudades_no_costeras.to_csv('./dataset/weatherAUS_LocationsCiudadesNoCosteras.csv',index=False)
dfLocations.to_csv('./dataset/weatherAUS_ForModel.csv',index=False)
```
